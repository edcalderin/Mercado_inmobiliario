### B칰squeda geogr치fica

Se calcular치n los establecimientos m치s cercanos a un punto dado, para ello se utilizar치 la distancia de Haversine o F칩rmula del semiverseno [https://en.wikipedia.org/wiki/Haversine_formula] :


\begin{align}
d = 2r\arcsin\left( \sqrt{\sin^{2} \left( \frac{\phi_2-\phi_1}{2} \right) + cos(\phi_1)cos(\phi_2)\sin^{2} \left( \frac{\lambda_2-\lambda_1}{2} \right)} \right)
\end{align}

Ejemplo de coordenadas (-34.6037389,-58.3837591) pertenecientes al Obelisco de Buenos Aires, obtenidas de https://www.google.com.ar/maps/place/Obelisco/@-34.6037389,-58.3837591,17z

Utilizando estas coordenadas encontramos los **2 departamentos** m치s cercanos con una distancia de **16 metros** aproximadamente.

```python
lat, lon = -34.6037389, -58.3837591
Utils.get_nearest_apartments(data, lat, lon)
```

#### Ajustando rango de valores para la variable 'surface_total' del dataframe

Calculando quantiles y Rango intercuant칤lico:

```python
surface_total = dataFiltered['surface_total']
Q1, Q3 = surface_total.quantile(.25), surface_total.quantile(.75)
IQR = Q3-Q1
```

```python
lowerLimit = Q1-1.5*IQR
upperLimit = Q3+1.5*IQR
```

```python
MIN_SURF, MAX_SURF = lowerLimit, upperLimit

Utils.plot_distribution(column_name='surface_total',
                        min_pivot=MIN_SURF, max_pivot=MAX_SURF, dataframe=dataFiltered)
```

* Se gener칩 un boxplot de superficie total por cada tipo de propiedad para observar la cantidad de outliers en cada una de estas clases.

* Esta variable presenta asimetr칤a y mas particularmente, una **asimetr칤a positiva** porque su Media > Mediana > Moda.

## 3. Machine Learning

* Creamos los datos de entrenamiento y validaci칩n utilizando las variables cuantitativas del dataframe filtrado.
* Para los modelos benchmark, se utilizar치 la m칠trica del error RMSE y calcular el que se ajuste mejor al problema.
* En la optimizaci칩n de hiperpar치metros, se utilizar치n las m칠tricas de precisi칩n y error con RMSE para determinar el mejor modelo, pero se har치 m치s 칠nfasis en la curva de errores.

```python
print(variables[:5])
```

```python
X = dataFiltered[variables[:5]]
y = dataFiltered[['price']]
```

Separando los datos de entrenamiento (70%) y validaci칩n (30%). Se activa la propiedad shuffle para no aleatorizar la selecci칩n:

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=.3, random_state=42, shuffle=False)
```

```python
dataFiltered.shape, X_train.shape, X_test.shape
```

### Modelo benchmark

Se crear치n modelos benchmarks para regresor lineal, 치rbol de decisi칩n y Vecinos mas cercanos con hiperpar치metros iniciales definidos de la siguiente manera:

* **Regresion lineal:** Sin par치metros.
* **Arbol de Decisi칩n:** Profundida de 4 y semilla pseudoaleatoria de 42.
* **Vecinos m치s Cercanos:** N칰mero de vecinos igul a a 5 y Distancia ponderada.

```python
modelos = ['Regresi칩n Lineal', '츼rbol de Decisi칩n', 'Vecinos m치s cercanos']

initialModelLinear = LinearRegression()
initialModelTree = DecisionTreeRegressor(max_depth=4, random_state=42)
initialModelKnn = KNeighborsRegressor(n_neighbors=5, weights='distance')
```

```python
initialModelLinear.fit(X_train, y_train)
initialModelTree.fit(X_train, y_train)
initialModelKnn.fit(X_train, y_train)
```

Se elegir치 como el mejor modelo benchmark aquel que presente el error RMSE m치s bajo:

```python
Utils.plot_errors_distribution(modelos, [
                               initialModelLinear, initialModelTree, initialModelKnn], X_train, y_train, X_test, y_test)
```

* Se obtuvieron menos errores con el modelo KNN.
* La distribuci칩n de errores es similar para entrenamiento y test, lo cual indica que no hay sobreajuste.
* Hay menores errores RMSE para KNN y la grafica de valores se ajusta m치s a la recta en este mismo modelo, quizas sea 칠ste el mejor modelo.
* Se utliz칩 la distancia ponderada weights='distance' con lo cual se obtuvo una reducci칩n bastante signficativa en los errores.
